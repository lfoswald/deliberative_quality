# Analysis of Comment Threads - Wildfires & Climate Change

```python
import pandas as pd
from datetime import datetime 
import networkx as nx 
import pickle
import pydot
from networkx.drawing.nx_pydot import graphviz_layout
import glob
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Load comment_df as pickle
with open('reddit_df.pickle', 'rb') as handle:
    reddit_df = pickle.load(handle)
```

```python
files = glob.glob('*.gexf')

all_threads = []

for graph in files:
    
    g = nx.read_gexf(graph)
    
    # geth info about nodes, edges, average in-degree and out-degree for all graphs
    print('\033[1m' + graph)
    print('\033[0m' + nx.info(g))
    print()
    
    threads = pd.Series([g.subgraph(c) for c in nx.weakly_connected_component_subgraphs(g)])
    
    subreddit_threads = []
    depths = []
    widths = []
    infos = []
    
    for c,g in enumerate(threads):
    
        comment_branch = list(nx.dag_longest_path(g))
        lpcb_df = reddit_df[reddit_df["id"].isin(comment_branch)]
        temp_sorted = lpcb_df.sort_values("depth",ascending=True)

        count = 0
        for index,row in temp_sorted.iterrows():
            if count < 1:
                c = row.parent_id                         
                subreddit_threads.append(c)
                count += 1
         
        # Thread info (level 0 comments)
        info = temp_sorted.loc[temp_sorted['depth'] == 0]
        infos.append(info)
        
        # Depths
        max_depth = len(comment_branch)
        depths.append(max_depth)
        
        # Widths
        mean_in_degrees = pd.Series(list(dict(g.in_degree()).values())).mean()
        widths.append(mean_in_degrees)
 
    # Quick Information               
    print('Number of discussion threads:' + '{}'.format(len(subreddit_threads)))
    print('Average depth: ' + '{}'.format(pd.Series(depths).mean()))
    print('Average width: ' + '{}'.format(pd.Series(widths).mean()))
    print('Maximum width: ' + '{}'.format(pd.Series(widths).max()))
    print('Maximum depth: ' + '{}'.format(pd.Series(depths).max()))
    print('\n')
    
    infos_df = pd.concat(infos)
    infos_df["depths"] = depths
    infos_df["widths"] = widths
    
    all_threads.append(infos_df)

all_threads_df = pd.concat(all_threads)
```

```python
# Store data (serialize)
with open('all_threads_df.pickle', 'wb') as handle:
    pickle.dump(all_threads_df, handle, protocol=pickle.HIGHEST_PROTOCOL)

# write to csv
all_threads_df.to_csv(r'all_threads_df.csv', index = True, header=True)
```

# Visualizations of comment graphs

```python
files = glob.glob('*.gexf')

graph_dict = {}

for graph in files:
    g = nx.read_gexf(graph)
    graph_dict[graph] = g
```

### Degree Distributions

```python
for graph, g in graph_dict.items():
    
    print('\033[1m' + graph)
    result = nx.degree(g) # Gather node degrees into a list
    deg_series = pd.Series(dict(result)) # Create degree series
    print('\033[0m' + "The average degree is {:.2f}.".format(deg_series.mean()))
    
    fig, ax = plt.subplots() 
    degplot = deg_series.hist(color='steelblue',grid=False); # display degree distribution
    degplot.set_xlabel("Degree", size = 16)
    #degplot.set_title(graph, size = 16)
    
    fig = matplotlib.pyplot.gcf()
    fig.set_size_inches(6,4)
    fig.savefig('degree_barplot_{}.png'.format(graph), dpi=200)
    
    plt.show()
    print()
```

### Weakly connected compontents

```python
for graph, g in graph_dict.items():
    
    wcc_climateskeptics = pd.Series([g.subgraph(c).number_of_nodes() for c in nx.weakly_connected_component_subgraphs(g)])
    
    fig, ax = plt.subplots() 
    wccplot = wcc_climateskeptics.plot(kind='hist',color='indianred')
    wccplot.set_xlabel("Number of nodes in wcc", size = 16)
    
    fig = matplotlib.pyplot.gcf()
    fig.set_size_inches(6,4)
    fig.savefig('wcc_barplot_{}.png'.format(graph), dpi=200)
    
    plt.show()
    print()
```

### Giant Component 

```python
# plot largest weakly connected component (giant component)
%pylab inline
import matplotlib.pyplot as plt 

for graph, g in graph_dict.items():
    plt.figure(graph)
    plt.title(graph, size=20)
    
    Gc = max(nx.weakly_connected_component_subgraphs(g), key=len)
    pos=nx.kamada_kawai_layout(Gc) 
    
    nx.draw_networkx_nodes(Gc,pos,node_size=40, node_color = "indianred")
    nx.draw_networkx_edges(Gc,pos)
    
    fig = matplotlib.pyplot.gcf()
    fig.set_size_inches(12,8)
    fig.savefig('Gc_{}.png'.format(graph), dpi=200)
    
    plt.show()
    print()
```

```python
# clean graph from isolates
def clean_graph(G,no_isolates=True):
    label_dict = {j:i for i,j in enumerate(G.nodes) }
    edge_set = set()
    for i in G.edges:
        il = list(i)
        il.sort()
        edge_set.add((label_dict[il[0]],label_dict[il[1]]))
    cleang = nx.DiGraph()
    cleang.add_nodes_from(label_dict.values())
    cleang.add_edges_from(edge_set)     
    # Now to remove the isolates
    if no_isolates: 
        cleang.remove_nodes_from(list(nx.isolates(cleang)))
    return cleang
```

```python
%pylab inline

for graph, g in graph_dict.items():
    
    ax = plt.subplots(figsize=(15, 15))
    
    pos = graphviz_layout(clean_graph(g), prog='dot')
    nx.draw(clean_graph(g), pos, node_size=40, node_color="blue", alpha=0.7, with_labels=False, arrows = False)
    
    plt.title(graph, size = 20)
    
    fig = matplotlib.pyplot.gcf()
    fig.set_size_inches(16,8)
    fig.savefig('comments_{}.png'.format(graph), dpi=200)
    
    plt.show()
    print()
```
